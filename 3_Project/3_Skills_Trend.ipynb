```python
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_de_DA = df[(df["job_country"] == "Germany") & (df["job_title_short"] == "Data Analyst")].copy()
df_de_DA.head()
```

```python
df_de_DA["job_posted_month_no"] = df_de_DA["job_posted_date"].dt.month
df_de_DA.head()
```

```python
df_de_DA.info()
```

```python
df_de_DA_exploded = df_de_DA.explode("job_skills")
df_de_DA_exploded.info()
```

```python
top_5_skills = df_de_DA_exploded.job_skills.value_counts().head().index.to_list()
top_5_skills
```

```python
df_de_DA_pivot = df_de_DA_exploded.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size")[top_5_skills]
df_de_DA_pivot
```

```python
df_de_DA_monthly_post = df_de_DA.groupby("job_posted_month_no").size()
df_de_DA_monthly_post
```

```python
df_de_DA_pivot = df_de_DA_pivot.div(df_de_DA_monthly_post/100, axis=0)
df_de_DA_pivot
```

```python
df_de_DA_pivot = df_de_DA_pivot.reset_index()
df_de_DA_pivot["job_posted_month"] = df_de_DA_pivot["job_posted_month_no"].apply(lambda month_no: pd.to_datetime(month_no, format="%m").strftime("%b"))
df_de_DA_pivot = df_de_DA_pivot.set_index("job_posted_month")
df_de_DA_pivot
```

```python
df_de_DA_pivot = df_de_DA_pivot.drop(columns=["job_posted_month_no"])
df_de_DA_pivot
```
