```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
```

```python
job_list = ["Data Scientist", "Data Engineer", "Data Analyst"]
#df_US = df[(df["job_title_short"].isin(job_list)) & (df["job_country"] == "United States")]
df_US = [df[(df["job_title_short"] == job_title) & (df["job_country"] == "United States")].salary_year_avg.value_counts() for job_title in job_list]
df_US
```

```python
df_US[0]
```

```python
plt.boxplot(df_US, job_list, vert=False)
```

```python
df
```

```python
df_US = df[df["job_country"] == "United States"]
df_US = df_US.dropna(subset="salary_year_avg")
df_US
```

```python
df_job_list = [df_US[df_US["job_title_short"] == job_title]["salary_year_avg"] for job_title in job_list]
df_job_list[0]
```

```python
plt.boxplot(df_job_list, tick_labels=job_list, vert=False)
plt.xlim((0, 600_000))
plt.xlabel("Salary Year Average")
plt.title("Salary Distribution in the US")

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x,pos: f"${int(x/1000)}K"))
ax.invert_yaxis()

plt.show()
```

```python
plt.boxplot([df.salary_year_avg.dropna()], tick_labels=["Salary year average"], vert=False)
plt.xlim((0, 600000))
plt.title("Salary distribution of all job posts in all countries")
plt.show()
```

```python
df_DS = df[(df["job_title_short"] == "Data Scientist") & df.salary_hour_avg.notna()].copy()
df_DS
```

```python
plt.boxplot(x=[df_DS.salary_hour_avg], tick_labels=["Data Scientist"], vert=False)
plt.title("Data Scientis Hourly Salary")
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f"${x}"))
plt.xlabel("Hourly salaries in USD")
```

```python
df_DA_US = df[(df["job_country"] == "United States") & (df["job_title_short"] == "Data Analyst")].copy()
df_DA_CD = df[(df["job_country"] == "Canada") & (df["job_title_short"] == "Data Analyst")].copy()
```

```python
df_DA_CD.dropna(subset=["salary_hour_avg"], inplace=True)
df_DA_US.dropna(subset=["salary_hour_avg"], inplace=True)
```

```python
plt.boxplot(x=[df_DA_CD.salary_hour_avg, df_DA_US.salary_hour_avg], tick_labels=["Data Analyst Job in Canada", "Data Analyst Job in USA"], vert=False)
plt.title("Data Analyst Hourly Salary in USD($) Comparison Between USA vs Canada")
plt.xlabel("Hourly Salary")
plt.xlim((0,250))
plt.show()
```
