```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_skills"] = df.job_skills.apply(lambda skills: ast.literal_eval(skills) if pd.notna(skills) else skills)
df.head()
```

```python
df = df.explode(column="job_skills")
df.head()
```

```python
df.job_skills.value_counts()
```

```python
df_job_skill_count = df.groupby(by=["job_title_short", "job_skills"]).size()
df_job_skill_count
```

```python
df_job_skill_count = df_job_skill_count.reset_index(name="skill_count")
df_job_skill_count
```

```python
df_job_skill_count.sort_values(by="skill_count", ascending=False, inplace=True)
df_job_skill_count
```

```python
job_title = "Cloud Engineer"
head_count = 15

df_job_skill_count_by_title = df_job_skill_count[df_job_skill_count["job_title_short"] == job_title].head(head_count)
df_job_skill_count_by_title.plot(kind="barh", x="job_skills", y="skill_count")

plt.gca().invert_yaxis()
plt.title(f"Top {head_count} skills for {job_title} position")
plt.xlabel("Job post count")
plt.legend().set_visible(False)
plt.show()
```

```python
df_avg_salary_skill = df.groupby(by="job_skills")["salary_year_avg"].median()
df_avg_salary_skill
```

```python
df_avg_salary_skill.sort_values(by="average_salary", ascending=False, inplace=True)
df_avg_salary_skill
```

```python
df_avg_salary_skill.set_index("job_skills", inplace=True)
```

```python
df_avg_salary_skill.head(10).plot(kind="barh", ylabel="job_skills", xlabel="average_salary", legend=False, title="Average Salary By Job Skill")
plt.gca().invert_yaxis()
plt.show()
```

```python
df_unique_skills = df.job_skills.value_counts()
df_unique_skills
```

```python
df_unique_skills = df_unique_skills.reset_index()
df_unique_skills["job_skills"] = df_unique_skills.job_skills.apply(lambda name: name.upper())
df_unique_skills.set_index("job_skills")
df_unique_skills.head(5).plot(kind="bar")
plt.xticks(rotation=45, ha="right")
plt.show()
```
