```python
def calculate_salary_based_on_seniority(dataframe):
  if "Senior" in dataframe["job_title_short"]:
    return 1.05 * dataframe["salary_year_avg"]
  else:
    return 1.03 * dataframe["salary_year_avg"]
```

```python
df_salary.salary_inflation_year_avg = df_salary.apply(calculate_salary_based_on_seniority, axis=1)
```

```python
def salary_category(dtf):
  match dtf["salary_rate"]:
    case "year":
      return "HIGH" if dtf["salary_year_avg"] > 100_000 else "AVERAGE"
    case "hour":
      return "HIGH" if dtf["salary_hour_avg"] > 60.0 else "AVERAGE"
```

```python
df_exercise["salary_category"] = df_exercise.apply(salary_category, axis=1)
df_exercise
```
